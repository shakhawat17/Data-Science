# Task 1: Basic Matplotlib (3 plots)


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gym = pd.read_csv("gym_membership.csv")
```

```python
print(gym.head())
print(gym.columns)
```

# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()

```python
plt.figure(figsize=(6,4))
gym['abonoment_type'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Gym Membership Types")
plt.xlabel("Membership Type")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])

```python
plt.figure(figsize=(6,4))
plt.hist(gym['avg_time_in_gym'], bins=10, color='orange', edgecolor='black')
plt.title("Distribution of Time Spent in Gym")
plt.xlabel("Average Time in Gym (hours)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])

```python
plt.figure(figsize=(6,4))
plt.scatter(gym['Age'], gym['visit_per_week'], color='green', alpha=0.6, edgecolors='black')
plt.title("Age vs Visits per Week")
plt.xlabel("Age")
plt.ylabel("Visits per Week")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=gym, x='gender', y='Age', palette="Set2")
plt.title("Age Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')


```python
plt.figure(figsize=(6,4))
sns.countplot(data=gym, x='days_per_week', palette="pastel")
plt.title("Gym Attendance Frequency (Days per Week)")
plt.xlabel("Days per Week")
plt.ylabel("Count")
plt.show()
```

# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')

```python
plt.figure(figsize=(6,4))
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week', palette="muted", estimator="mean", ci=None)
plt.title("Average Visits per Week by Membership Type")
plt.xlabel("Membership Type")
plt.ylabel("Average Visits per Week")
plt.show()
```
